```python
df_z_scaled = df.copy()
  
# apply normalization technique to Column 1
for column in list(df_z_scaled.columns):
    df_z_scaled[column] = (df_z_scaled[column] - df_z_scaled[column].mean()) / df_z_scaled[column].std()    

# view normalized data  
display(df_z_scaled)
```
